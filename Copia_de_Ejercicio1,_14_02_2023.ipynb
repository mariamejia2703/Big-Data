```python
import numpy as np #manejo de listas (vectores) [a,b,c,d]
import pandas as pd #manejo de DataFrame
import matplotlib.pyplot as plt
import seaborn as sns #visualizaciones
```

importamos la base de datos

```python
df = pd.read_csv("/content/drive/MyDrive/bigdata/StudentsPerformance.csv") #agg base de datos
df.head()
```

```python
df.info()#ver atributos de la base de datos
```

```python
df.describe()#informacion estadistica de la base de datos
```

```python
df = df.rename(columns={'parental level of education':'parental_education'})#renombrar columnas 
```

```python
df = df.rename(columns={'test preparation course':'preparation'})
```

```python
df.info()
```

```python
df['gender']=df['gender'].astype('category') #renombrar variables categoricas para poder entrar a consultar
df['race/ethnicity']=df['race/ethnicity'].astype('category')
df['parental_education']=df['parental_education'].astype('category')
df['lunch']=df['lunch'].astype('category')
df['preparation']=df['preparation'].astype('category')
```

```python
df.info()
```

```python
df['gender'].value_counts() #ver cuantos datos hay de esta categoria en el dataframe
```

```python
df['race/ethnicity'].value_counts()
```

```python
sns.countplot(data=df,x='gender') #diagrama de columnas, se hace cuando tenemos variables categoricas.
```

```python
sns.histplot(data=df,x='math score',bins=10) #histograma, solo variables categoricas, bins es la canttidad de barras que queremos dividir.
```

```python
sns.histplot(data=df,x='writing score', bins=10)
```

```python
sns.histplot(data=df, x='reading score', bins=10)
```

```python
fig=plt.figure(figsize=(20,7))
ax1=fig.add_subplot(131)#definir las variables para distinguir las columnas
ax2=fig.add_subplot(132)
ax3=fig.add_subplot(133)

#sns.boxplot(data=df,y='math score')#diagrama de cajas
sns.boxplot(data=df,y='reading score',ax=ax1)#separar las columnas en diagrama de cajas.
sns.boxplot(data=df,y='math score',ax=ax2)
sns.boxplot(data=df,y='writing score',ax=ax3)
```
